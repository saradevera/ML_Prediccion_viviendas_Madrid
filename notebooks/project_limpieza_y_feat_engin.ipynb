```python
# ESTUDIAMOS CADA VARIABLE DE MANERA INDEPENDIENTE PARA TRATAR LOS NaN DE LA MANERA QUE CORRESPONDA. 

## EN MUCHOS CASOS, TENEMOS TRUE Y NANS, POR LO QUE DEDUCIMOS QUE LOS NAN CORRESPONDEN A FALSE. 
## EN LOS CASOS EN LOS QUE NO SE CUMPLE ESO, LO SUSTITUIMOS CON UN CRITERIO DISTINTO DEPENDIENDO DE CADA CASO

# df.dtypes
# for i in df:
#     print(i)
#     print(df[i].value_counts())
#     print('')
```

```python
# COMENZAMOS A SUSTITUIR LOS VALORES NAN, CON DISTINTOS CRITERIOS

lista = ['has_garden','has_balcony','has_green_zones','has_pool','has_storage_room','has_terrace','has_ac','has_ac','has_fitted_wardrobes',
'has_central_heating','has_individual_heating','has_lift','has_parking',
'is_accessible','is_floor_under','is_new_development',
'is_orientation_west','is_orientation_east','is_orientation_north','is_orientation_south']

# SUSTITUIMOS LOS NAN POR FALSE EN ESTAS COLUMNAS, YA QUE HEMOS VISTO QUE SOLO TIENEN VALORES TRUE O NAN, POR LO QUE ENTENDEMOS QUE LOS NAN SON FALSE
for i in lista:
    df[i].fillna(False, inplace=True)

# A CONTINUACIÓN LO TRANSFORMAMOS A 0 Y 1
for i in lista:
    df[i]= df[i].map(lambda x: 0 if x == False  else 1)
```

```python
# SUSTITUIMOS EL RESTO DE NANS CADA UNO CON SU CRITERIO

df['is_exterior'].fillna(True, inplace=True)

df['parking_price'].fillna(0, inplace=True)

df['built_year'].fillna(df['built_year'].median(), inplace=True)

df['floor'].fillna(1, inplace=True)

df['house_type_id'].fillna('HouseType 1: Pisos', inplace=True)

df['sq_mt_built'].fillna(df['sq_mt_built'].mean(), inplace=True)

df['n_bathrooms'].fillna(df['n_bathrooms'].mean(), inplace=True)

# df['buy_price_with_park'] = df['buy_price'] + df['parking_price']
```

```python
# COMENZAMOS A TRATAR LAS VARIABLES CATEGÓRICAS, PARA CONVERTIRLAS EN VARIABLES NUMÉRICAS

print(df['house_type_id'].value_counts())

dict_housetype = {'HouseType 1: Pisos':1,
                'HouseType 2: Casa o chalet':2,
                'HouseType 5: Áticos':3,
                'HouseType 4: Dúplex':4,
                # 'HouseType 0: Otros':0
                }

df['house_type_id'] = df['house_type_id'].replace(dict_housetype)
df['house_type_id']= df['house_type_id'].apply(lambda x: float(x))
```

```python
df.groupby(['house_type_id'])[['buy_price']].sum()
```

```python
house_type_id = df.pop('house_type_id')

df['Pisos'] = (house_type_id==1)*1
df['Casa o chalet'] = (house_type_id==2)*1
df['Áticos'] = (house_type_id==3)*1
df['Dúplex'] = (house_type_id==4)*1
# df['Otros '] = (house_type_id==0)*1
```

```python
# SUSTITUIMOS EL RESTO DE VARIABLES BOOLEANAS POR 0 Y 1

df['is_floor_under'] = df['is_floor_under'].map(lambda x: 0 if x == False  else 1)
df['is_renewal_needed'] = df['is_renewal_needed'].map(lambda x: 0 if x == False  else 1)
df['is_new_development'] = df['is_new_development'].map(lambda x: 0 if x == False  else 1)
df['is_exterior'] = df['is_exterior'].map(lambda x: 0 if x == False  else 1)
```

```python
# TRANSFORMAMOS EN int LOS DATOS CATEGÓRICOS DE LA VARIABLE FLOOR

list_planta_baja=['Bajo','Entreplanta exterior','Entreplanta','Entreplanta interior']
list_planta_sot=['Semi-sótano exterior','Semi-sótano interior','Semi-sótano','Sótano interior','Sótano','Sótano exterior']

for i in list_planta_baja:
    df[df['floor']==i] = df[df['floor']==i].apply(lambda x: x.replace(i,0.5))

for i in list_planta_sot:
    df[df['floor']==i] = df[df['floor']==i].apply(lambda x: x.replace(i,0))

# for i in list_planta_baja:
    # df[df['floor']==i]=0.5

# for i in list_planta_sot:
#     df[df['floor']==i]=0

df['floor']= df['floor'].astype(float)
```

```python
# VAMOS A TRABAJAR CON LA VARIABLE CATEGÓRICA DE 'neighborhood_id'

df['distrito_name'] = df['neighborhood_id'].str.extract(r'(District \d+)')
df['distrito_id_num'] = df['distrito_name'].str.extract(r'(\d+)')
df['distrito_id_num']= df['distrito_id_num'].astype(float)
```

```python
df['neighborhood_id_num'] = df['neighborhood_id'].str.extract(r'(\d+)')
df['neighborhood_id_num']= df['neighborhood_id_num'].apply(lambda x: float(x))
```

```python
df['neighborhood_price_m2'] = df['neighborhood_id'].str.extract(r'(\d+\.\d+)')
df['neighborhood_price_m2']= df['neighborhood_price_m2'].apply(lambda x: float(x))

df['neighborhood_price_m2'].fillna(df['neighborhood_price_m2'].mean(), inplace=True)
```

```python
# LA COLUMNA DE buy_price_by_area ESTA SACADA DE LA COLUMNA buy_price/sq_mt_built POR LO QUE NO NOS SERVIRÍA PARA DATOS NUEVOS YA QUE NO TENDRÍAMOS
# PRECISAMENTE LA COLUMNA DE buy_price (ES EL TARGET).

df[['buy_price','buy_price_by_area','sq_mt_built','neighborhood_price_m2']]
```

```python
# VAMOS A CREAR UNA COLUMNA NUEVA MULTIPLICANDO LA COLUMNA sq_mt_built POR LA COLUMNA QUE HEMOS OBTENIDO DE neighborhood_price_m2 YA QUE AMBOS DATOS LOS
# CONOCEREMOS EN CONSULTAS NUEVAS

df['estimado_metros_por_preciom2'] = df['sq_mt_built']*df['neighborhood_price_m2']
```

```python
# TRATAMOS DE NUEVO DUPLICADOS POR SI SE HUBIERAN FORMADO AL TRANSFORMAR Y ELIMINAR COLUMNAS

# df.drop_duplicates(['subtitle','floor','sq_mt_built','n_rooms','n_bathrooms','buy_price'], keep='first', inplace=True)

# ELIMINAMOS LAS COLUMNAS QUE NO VAMOS A NECESITAR NI PARA EL EDA NI PARA ENTRENAMIENTO DE MODELOS

df.drop(columns=['subtitle', 'neighborhood_id', 'distrito_name', 'buy_price_by_area'], inplace=True)
```
